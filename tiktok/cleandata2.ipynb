```python

acounts = pd.read_csv("./tiktokdataset/accounts.txt",sep='\t')

videos = pd.read_csv("./tiktokdataset/videos.txt",sep='\t')
acounts.columns
videos.columns
```

```python
# acounts = acounts.dropna(subset=['IsABot'])
# acounts["IsABot"].value_counts(dropna=False)
accounts=acounts.loc[:,["Id","IsVerified","HasProfilePicture","AnzahlFolgeIch","AnzahlFollower",'HasAccountDescription','AnzahlLikes','AnzahlPosts','AverageNumberOfHashtags','AverageNumberOfComments','AverageNumberOfForwardings','AverageNumberOfLikes','AverageNumberOfLinkedProfiles','AverageNumberOfViews','IsABot']]
accounts=accounts.rename(columns={"AnzahlFolgeIch":'following','AnzahlFollower':'follower','AnzahlLikes':'likes','AnzahlPosts':'posts','AverageNumberOfForwardings':'AverageNumberOfShare'})
accounts
```

```python
accounts=accounts.drop_duplicates(subset="Id",keep='first')
```

```python
accounts['avg_videos_by_day_of_week']=0
```

```python
videos.iloc[15865]
```

```python
def date(year):
    try:
        dt = datetime.datetime.strptime(str(year), '%Y-%m-%d')
        return dt
    except  Exception as e:
        return 0

for j,ac in accounts.iterrows():
    monday_to_sunday = [0] * 7
    v=videos[videos['AccountId']==ac['Id']]
    v=pd.DataFrame(v)
   
    if(len(v)>= 1):
        for x,i in v.iterrows():
            if pd.isna(i["DatumVideoUpload"]):
                print("true")
                continue
            vi_time = date(i["DatumVideoUpload"])
            # print((i["DatumVideoUpload"])," ",date(i["DatumVideoUpload"]))
           
            monday_to_sunday[vi_time.weekday()] += 1
        avg_videos_by_day_of_week = round(
                sum(monday_to_sunday)/len(monday_to_sunday), 3)
        
        accounts.loc[j, 'avg_videos_by_day_of_week'] = avg_videos_by_day_of_week
        
    else:
        pass
```

```python
accounts.to_csv('./Data/cleandata1.csv', index=False)
```

```python

# assume your DataFrame is named 'df'
import numpy as np
# replace 'True' with 1 and 'False' with 0
df = accounts.replace({'True': 1, 'False': 0})
df = df.replace({True: 1, False: 0})
# replace all non-numeric string values with -1
df = df.apply(pd.to_numeric, errors='coerce').fillna(np.nan)
df['IsABot'] = df['IsABot'].replace(np.nan, -1)
# drop rows that contain a NaN value in any column
df = df.dropna()
```

```python
df = df.astype(float).round(2)

cols_to_modify = df.columns.drop('IsABot')

df[cols_to_modify] = df[cols_to_modify].abs()
```

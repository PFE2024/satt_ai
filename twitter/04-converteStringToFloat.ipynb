The Levenshtein distance is a measure of the minimum number of single-character edits (insertions, deletions, or substitutions)

```python
def levenshtein_distance(s1, s2):
    m, n = len(s1), len(s2)
    if m < n:
        return levenshtein_distance(s2, s1)
    if n == 0:
        return m

    previous_row = range(n+1)
    for i, c1 in enumerate(s1):
        current_row = [i+1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j+1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row

    return previous_row[n]
```

```python
data["display_name"] = data["display_name"].fillna('')

data.loc[:,'userNameScore']= data.apply(lambda row : 1 - (levenshtein_distance(row["screen_name"],	row["display_name"]) / max(len(row["screen_name"]), len(row["display_name"]))), axis=1)
data
```

converte date to only years

```python
def creation_year(year):
    try:
        dt = datetime.strptime(str(year), '%Y-%m-%d %H:%M:%S')
        return dt.year
    except:
        return 0
data['date_joined'] = data.apply(lambda row :creation_year(row["date_joined"]), axis=1)

data['most_recent_post'] = data.apply(lambda row :creation_year(row["most_recent_post"]), axis=1)
data =data.drop(["display_name","background_image"], axis=1)
data
```
